## EDA

```python
df = pd.read_csv("Online Retail.csv",encoding = 'latin1')
df.head()
```

```python
## change in column name 
df.rename(columns={"ï»¿InvoiceNo":"InvoiceNO"},inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Handeling the missing values 

df = df[df['Quantity']>0]
df = df.dropna(subset=['CustomerID'])
```

```python
df.head()
```

```python
basket = (df
          .groupby(['InvoiceNO', 'Description'])['Quantity']
          .sum()
          .unstack()
          .fillna(0))

basket = basket.applymap(lambda x: 1 if x > 0 else 0)
```

```python
frequent_items = apriori(basket,
                         min_support=0.02,
                         use_colnames=True)

print(frequent_items.head())
```

```python
rules = association_rules(frequent_items,
                          metric="lift",
                          min_threshold=1)

print(rules.head())
```

```python
frequent_items.to_csv("frequent_items.csv")
rules.to_csv("association_rules.csv")
```
